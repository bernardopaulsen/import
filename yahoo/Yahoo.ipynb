```python
import matplotlib.pyplot as plt
import yahoo
```

```python
start = (2015,1,1)
end   = (2020,1,1)
tickers = ['BBAS3.SA','CESP6.SA','JBSS3.SA','IGTA3.SA','EZTC3.SA','LAME4.SA']
adj_close = yahoo.multiple(tickers,'Adj Close',start,end)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
for ticker in adj_close.columns:
    ax.plot(adj_close[ticker],label=ticker)
ax.legend()
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.savefig('adj_close')
plt.show()
```

```python
ret = yahoo.multiple(tickers,'Ret',start,end)
ret = ret.dropna()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
for ticker in ret.columns:
    ax.plot(ret[ticker],label=ticker)
ax.legend()
plt.xlabel('Date')
plt.ylabel('Return')
plt.savefig('ret')
plt.show()
```

```python
foreign = ['AAPL','ADBE','FB','KO','EBAY','EA']
foreign_adj_close = yahoo.multiple(foreign,'Adj Close',start,end)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
for ticker in foreign_adj_close.columns:
    ax.plot(foreign_adj_close[ticker],label=ticker)
ax.legend()
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.savefig('for_adj_close')
plt.show()
```
