```python
import bacen
import matplotlib.pyplot as plt
```

```python
di = bacen.get(12,(2000,1,1),(2020,1,1)).valor
di.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(di)
plt.xlabel('Date')
plt.ylabel('DI')
plt.savefig('di')
plt.show()
```

```python
dol = bacen.get(1,(2000,1,1),(2020,1,1)).valor
dol.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(dol)
plt.xlabel('Date')
plt.ylabel('BRL/USD')
plt.savefig('dol')
plt.show()
```
